```python
with open("rosalind_prot.txt") as f:
    mRNA = f.read().split()

mRNA = "".join(mRNA)
codon_size = 3

RNA_string_splited = [mRNA[codon:codon + codon_size] for codon in range(0, len(mRNA), codon_size)]

intermediate_protein = [RNA_codon_table[codon] for codon in RNA_string_splited if RNA_codon_table[codon] != "Stop" ]
final_protein = "".join(intermediate_protein)

final_protein
```
